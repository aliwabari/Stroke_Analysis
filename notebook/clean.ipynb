```python
import pandas as pd 
```

```python
df = pd.read_csv("../data/raw/stroke_data.csv")
df.head()

df_clean= df.copy()
```

```python
#Handle null values in bmi column 
df_clean=df_clean.fillna(df_clean['bmi'].median())
df_clean.isnull().sum()
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df_clean["bmi_category"] = df_clean["bmi"].apply(bmi_category)
```

```python
df_clean.to_csv("../data/clean/clean_data.csv" , index=False)
```
