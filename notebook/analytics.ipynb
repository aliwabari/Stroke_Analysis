```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
df = pd.read_csv("../data/clean/clean_data.csv")
df.head()
```

```python
(round(df.groupby("smoking_status")["stroke"].mean().sort_values(ascending=False )* 100,1))
```

```python
bmi_levels = round(df.groupby("bmi_category")["stroke"].mean().sort_values(ascending=False) * 100, 1)

plt.figure()
plt.pie(bmi_levels.values, labels=bmi_levels.index, autopct='%1.1f%%')
plt.title('BMI levels')
plt.show()
```

```python
Married=(round(df.groupby("ever_married")["stroke"].mean().sort_values(ascending=False )* 100,1))
plt.figure()
plt.pie(Married.values, labels=Married.index, autopct='%1.1f%%')
plt.title('Gender')
plt.show()
```

```python

res = round(df.groupby('Residence_type')['stroke'].mean() * 100, 1)

plt.bar(res.index, res.values)
plt.xlabel('Residence Type')
plt.ylabel('Stroke Percentage')
plt.title('Stroke Rate by Residence Type')
plt.show()
```

```python
heart_disease = round(df.groupby('heart_disease')['stroke'].mean() * 100, 1)

plt.figure()
plt.bar(['No', 'Yes'], heart_disease.values)
plt.xlabel('Heart Disease')
plt.ylabel('Stroke Percentage')
plt.title('Stroke Rate by Heart Disease')
plt.show()
```
